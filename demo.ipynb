```python
url = 'http://localhost:5000/comments?'
response = requests.get(url+'collection')
response.json()[:10]
```

```python
url = 'http://localhost:5000/'
params = {'game' : 'The Legend of Zelda: Ocarina of Time'}
response = requests.get(url+'comments',params=params)
response.json()[:2]
```

```python
url = 'http://localhost:5000/'
params = {'game' : 'The Legend of Zelda: Ocarina of Time', 'limit' : 2}
response = requests.get(url+'platforms',params=params)
response.json()
```

```python
url = 'http://localhost:5000/'
comment = {
    'game' : 'The Legend of Zelda: Ocarina of Time',
    'platform' : 'PSP',
    'username': 'sweetbabyjezuz',
    'comment' : 'i like this game'
}
response = requests.get(url+'newcomment',data=comment)
response.text
```
